```python
tf_idf = pd.read_csv('tf_idf_matrix.csv')
tf_idf.rename(columns={
    'Unnamed: 0': 'location'
}, inplace=True)
```

```python
# 각 location 별로 tf-idf 높게 나온 term을 뽑아보자!

terms_dict = dict()

for i, row in tf_idf.iterrows():
    vectors = dict()
    for term in row.keys():
        if term == 'location':
            continue
        vectors[term] = row[term]
    terms_dict[row['location']] = vectors
```

```python
tf_idf_dict = dict()

for location, terms in terms_dict.items():
    tf_idf_dict[location] = sorted(terms.items(), key=lambda x: x[1], reverse=True)[:20]
    
    
    
```

```python
tf_idf_dict
```

```python
# 각 지역의 대표 해시태그들?
from collections import defaultdict
import re 

hashtag_count_dict = dict()

def str2list(list_str):
    return [tag for tag in re.split("\[|\]|\'|\,| ",list_str) if len(tag) > 0]

new_clr = clr.copy()
new_clr['hashtags']  = clr['hashtags'].map(lambda x : str2list(x))


for i, row in new_clr.iterrows():
    location = row['location']
    hashtags = row['hashtags']
    hashtag_dict = defaultdict(int)
    
    for hashtag in hashtags:
        for term, value in tf_idf_dict[location]:
            if term in hashtag:
                hashtag_dict[hashtag] += 1
    hashtag_count_dict[location] = hashtag_dict

tags_rows = []

for location, hashtags in hashtag_count_dict.items():
    print(location)
    top_hashtags = sorted(hashtags.items(), key=lambda x: x[1], reverse=True)
    top_hashtags = [(hashtag.replace('제주도', '').replace('제주', ''), count) for hashtag, count in top_hashtags]
    print(top_hashtags)
    print()
    
    text = ''
    
    for hashtag, count in top_hashtags:
        text += ' ' + ' '.join([hashtag] * count)
    
    row = {
        'location': location,
        'text': text
    }
    tags_rows.append(row)
    
tags_df = pd.DataFrame(tags_rows)
tags_df
```

```python
locations = pd.read_csv('dataframes/cluster_features_filtered.csv')
locations.sort_values(by='count_hashtags', inplace=True, ascending=False)
```

```python
locations
```

```python
cluster_result = pd.merge(locations, memb, on="location")
cluster_result
```

```python
cluster_hashtags_rows = []

def merge_lists(lists):
    merged_list = []
    for l in lists:
        l = str2list(l)
        merged_list.extend(l)
        
    return merged_list

for cluster, group in cluster_result.groupby('membership'):
    cluster_hashtags_rows.append({
        'cluster': cluster,
        'locations': list(group['location']),
        'hashtags': merge_lists(group['hashtags'])
    })
    
cluster_hashtags = pd.DataFrame(cluster_hashtags_rows)
cluster_hashtags
```

```python
from itertools import chain

cluster_hashtags_dict = dict()
cluster_locations_dict = dict()

for i, row in cluster_hashtags.iterrows():
    cluster = row['cluster']
    locations = row['locations']
    hashtags = row['hashtags']
    hashtag_dict = defaultdict(int)
    
    cluster_locations_dict[cluster] = locations
    
    for hashtag in hashtags:
        for term, value in chain(*[tf_idf_dict[location] for location in locations]):
            if term in hashtag:
                hashtag_dict[hashtag] += 1
    cluster_hashtags_dict[cluster] = sorted(hashtag_dict.items(), key=lambda x: x[1], reverse=True)
    
cluster_hashtags_dict
```

```python
cluster_repr_hashtag_rows = []

for cluster in cluster_hashtags_dict:
    selected_hashtags = []
    count = 0
    for tag, value in cluster_hashtags_dict[cluster]:
        duplicate = False
        for prev_tag in selected_hashtags:
            if (prev_tag in tag) or (tag in prev_tag):
                duplicate = True
                break
        if duplicate:
            continue
        selected_hashtags.append(tag)
        count += 1
        
        if count == 5:
            break
            
    cluster_repr_hashtag_rows.append({
        'cluster': cluster,
        'locations': cluster_locations_dict[cluster],
        'hashtags': selected_hashtags
    })
    
cluster_repr_hashtag = pd.DataFrame(cluster_repr_hashtag_rows)
cluster_repr_hashtag.to_csv('cluster_repr_hashtag.csv')
```

```python
places = list(set(tags_cluster['place']))
places
```

```python
cluster_theme_rows = []

for cluster, group in tags_cluster.groupby('membership'):
    row = {
        'cluster': cluster,
        'locations': cluster_locations_dict[cluster]
    }
    for place in places:
        row[place] = sum(group[group['place'] == place]['count'])
    cluster_theme_rows.append(row)
    
cluster_themes = pd.DataFrame(cluster_theme_rows)
```
