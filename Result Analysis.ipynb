```python
tf_idf = pd.read_csv('tf_idf_matrix.csv')
tf_idf.rename(columns={
    'Unnamed: 0': 'location'
}, inplace=True)
```

```python
# 각 location 별로 tf-idf 높게 나온 term을 뽑아보자!

terms_dict = dict()

for i, row in tf_idf.iterrows():
    vectors = dict()
    for term in row.keys():
        if term == 'location':
            continue
        vectors[term] = row[term]
    terms_dict[row['location']] = vectors
```

```python
tf_idf_dict = dict()

for location, terms in terms_dict.items():
    tf_idf_dict[location] = sorted(terms.items(), key=lambda x: x[1], reverse=True)[:20]
    
    
    
```

```python
tf_idf_dict
```

```python
# 각 지역의 대표 해시태그들?
from collections import defaultdict
import re 

hashtag_count_dict = dict()

def str2list(list_str):
    return [tag for tag in re.split("\[|\]|\'|\,| ",list_str) if len(tag) > 0]

new_clr = clr.copy()
new_clr['hashtags']  = clr['hashtags'].map(lambda x : str2list(x))


for i, row in new_clr.iterrows():
    location = row['location']
    hashtags = row['hashtags']
    hashtag_dict = defaultdict(int)
    
    for hashtag in hashtags:
        for term, value in tf_idf_dict[location]:
            if term in hashtag:
                hashtag_dict[hashtag] += 1
    hashtag_count_dict[location] = hashtag_dict

tags_rows = []

for location, hashtags in hashtag_count_dict.items():
    print(location)
    top_hashtags = sorted(hashtags.items(), key=lambda x: x[1], reverse=True)
    top_hashtags = [(hashtag.replace('제주도', '').replace('제주', ''), count) for hashtag, count in top_hashtags]
    print(top_hashtags)
    print()
    
    text = ''
    
    for hashtag, count in top_hashtags:
        text += ' ' + ' '.join([hashtag] * count)
    
    row = {
        'location': location,
        'text': text
    }
    tags_rows.append(row)
    
tags_df = pd.DataFrame(tags_rows)
tags_df
```
