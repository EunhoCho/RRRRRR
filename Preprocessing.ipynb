```python
import math

def filter_posts_by_hashtag(posts_df, hashtag_patterns):
    rows = []
    for i, post in posts_df.iterrows():
        if not 'hashtags' in post.keys():
            continue
        
        hashtags = post['hashtags']
        
        if type(hashtags) == float and math.isnan(hashtags):
            continue
        
        contains_pattern = False
        for hashtag in hashtags:
            for pattern in hashtag_patterns:
                if pattern in hashtag:
                    contains_pattern = True
                    break
            
            if contains_pattern:
                break
                
        if contains_pattern:
            rows.append(post)
            
    return pd.DataFrame(rows)
```

```python
def flatten_hashtags(posts):
    rows = []
    for post in posts:
        if not 'hashtags' in post.keys():
            continue
        
        hashtags = post['hashtags']
            
        for hashtag in hashtags:
            rows.append({
                'key': post['key'],
                'location': post['location'],
                'datetime': post['datetime'],
                'hashtag': hashtag
            })
            
    return pd.DataFrame(rows)
```

## Getting x, y coordinates from location

```python
import importlib
import loc2coord
importlib.reload(loc2coord)

from loc2coord.main import getLatLng

def convert_location_to_coord(df):
    rows = []
    total = len(df)
    count = 1
    for i, row in df.iterrows():
        print(f'processing {count}/{total}')
        count += 1
        
        loc = getLatLng(row['location'])
        if loc == None:
            continue
        
        print(f'Finding coordinates: {loc}')
        rows.append({
            'key': row['key'],
            'loc_x': loc['x'],
            'loc_y': loc['y'],
            'address': loc['address'],
            'datetime': row['datetime'],
            'hashtag': row['hashtag']
        })
    return pd.DataFrame(rows)

df_loc = convert_location_to_coord(df)
```

## Classify with hashtag category

```python
posts_df = pd.read_csv('dataframes/posts.csv')
df_loc = pd.read_csv('dataframes/locations_coord.csv')

addresses = list(set(df_loc['address']))

def address2unit(address):
    """
    '읍', 면', '동' 단위의 주소만 뽑아내기
    """
    units = address.split()
    target = []
    for unit in units:
        if unit[-1] == '읍':
            target.append(unit)
        if unit[-1] == '면':
            target.append(unit)
        if unit[-1] == '동':
            processed_unit = unit
            if unit[-2] == '일':
                processed_unit = unit[:-2] + '1' + '동'
            if unit[-2] == '이':
                processed_unit = unit[:-2] + '2' + '동'
            if unit[-2] == '삼':
                processed_unit = unit[:-2] + '3' + '동'
            target.append(processed_unit)
            
    assert len(target) == 1
    return target[0]

address_units = [address2unit(address) for address in addresses]

address_units = list(set(address_units))
```

```python
new_rows = []
for i, row in df_loc.iterrows():
    row['address_unit'] = address2unit(row['address'])
    new_rows.append(row)
    
df_loc = pd.DataFrame(new_rows)
df_loc.to_csv('dataframes/locations_coord.csv')
```

```python
from collections import defaultdict

locations = address_units
places = ['맛집', '카페', '술집', '한달살기']
places_tags = {
    '맛집': ['맛집', '식당', '맛', '밥'],
    '카페': ['카페', '커피', 'cafe'],
    '술집': ['술', '술집', 'pub', '펍', '맥주', '와인', 'bar'],
    '한달살기': ['한달']
}

def is_match_tag(tag, place_key):
    for place in places_tags[place_key]:
        if place in tag:
            return True
        
    return False

def generate_location_tags(hashtags, locations, places):
    hashtag_dict = defaultdict(int)
    
    for i, row in df_loc.iterrows():
        hashtag = row['hashtag']
        tag_type = [None, None]
        
        for location in locations:
            if location in row['address']:
                tag_type[0] = location
                
        for place in places:
            if is_match_tag(hashtag, place):
                tag_type[1] = place
                
        if not None in tag_type:
            hashtag_dict[tuple(tag_type)] += 1
        
    rows = []
    for location, place in hashtag_dict:
        rows.append({
            'location': location,
            'place': place,
            'count': hashtag_dict[(location, place)]
        })
        
    return pd.DataFrame(rows)

location_tags = generate_location_tags(hashtag_df, locations, places)
location_tags.to_csv('dataframes/location_tags.csv')
```
