```python
import math

def filter_posts_by_hashtag(posts_df, hashtag_patterns):
    rows = []
    for i, post in posts_df.iterrows():
        if not 'hashtags' in post.keys():
            continue
        
        hashtags = post['hashtags']
        
        if type(hashtags) == float and math.isnan(hashtags):
            continue
        
        contains_pattern = False
        for hashtag in hashtags:
            for pattern in hashtag_patterns:
                if pattern in hashtag:
                    contains_pattern = True
                    break
            
            if contains_pattern:
                break
                
        if contains_pattern:
            rows.append(post)
            
    return pd.DataFrame(rows)
```

```python
def flatten_hashtags(posts):
    rows = []
    for post in posts:
        if not 'hashtags' in post.keys():
            continue
        
        hashtags = post['hashtags']
            
        for hashtag in hashtags:
            rows.append({
                'key': post['key'],
                'location': post['location'],
                'datetime': post['datetime'],
                'hashtag': hashtag,
                'address': post['address'],
                'loc_x': post['x'],
                'loc_y': post['y']
                
            })
            
    return pd.DataFrame(rows)
```

```python
posts_df = pd.read_csv('dataframes/posts.csv')


addresses = list(set(df_loc['address']))

def address2unit(address):
    """
    'ì', ë©´', 'ë™' ë‹¨ìœ„ì˜ ì£¼ì†Œë§Œ ë½‘ì•„ë‚´ê¸°
    """
    units = address.split()
    target = []
    for unit in units:
        if unit[-1] == 'ì':
            target.append(unit)
        if unit[-1] == 'ë©´':
            target.append(unit)
        if unit[-1] == 'ë™':
            processed_unit = unit
            if unit[-2] == 'ì¼':
                processed_unit = unit[:-2] + '1' + 'ë™'
            if unit[-2] == 'ì´':
                processed_unit = unit[:-2] + '2' + 'ë™'
            if unit[-2] == 'ì‚¼':
                processed_unit = unit[:-2] + '3' + 'ë™'
            target.append(processed_unit)
            
    assert len(target) == 1
    return target[0]

address_units = [address2unit(address) for address in addresses]

address_units = list(set(address_units))
```

```python
from collections import defaultdict

locations = address_units
places = ['ë§›ì§‘', 'ì¹´í˜', 'ìˆ ì§‘', 'ìˆ™ë°•']
places_tags = {
    'ë§›ì§‘': ['ë§›ì§‘', 'ì‹ë‹¹', 'ë§›', 'ë°¥'],
    'ì¹´í˜': ['ì¹´í˜', 'ì»¤í”¼', 'cafe'],
    'ìˆ ì§‘': ['ìˆ ', 'ìˆ ì§‘', 'pub', 'í', 'ë§¥ì£¼', 'ì™€ì¸', 'bar'],
    'ìˆ™ë°•': ['í•œë‹¬', 'ì‚´ê¸°', 'ìˆ™ë°•', 'íœì…˜', 'í˜¸í…”', 'ë¯¼ë°•', 'ìˆ™ì†Œ']
}

def is_match_tag(tag, place_key):
    for place in places_tags[place_key]:
        if place in tag:
            return True
        
    return False

def generate_location_tags(hashtags, locations, places):
    hashtag_dict = defaultdict(int)
    
    for i, row in df_loc.iterrows():
        hashtag = row['hashtag']
        tag_type = [None, None]
        
        for location in locations:
            if location in row['address']:
                tag_type[0] = location
                
        for place in places:
            if is_match_tag(hashtag, place):
                tag_type[1] = place
                
        if not None in tag_type:
            hashtag_dict[tuple(tag_type)] += 1
        
    rows = []
    for location, place in hashtag_dict:
        rows.append({
            'location': location,
            'place': place,
            'count': hashtag_dict[(location, place)]
        })
        
    return pd.DataFrame(rows)

location_tags = generate_location_tags(hashtag_df, locations, places)
location_tags.to_csv('dataframes/location_tags.csv')
```
